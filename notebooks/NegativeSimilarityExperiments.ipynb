```python
pos = pd.read_pickle('../data/training/positive/mibig_bgcs_all.domain_set.pkl')
```

```python
cols = np.intersect1d(neg.columns, pos.columns)
neg = neg[cols]
pos = pos[cols]
```

```python
len(neg)
```

```python
pos.head()
```

```python
distances = pd.DataFrame(cosine_similarity(neg.values, pos.values), index=neg.index, columns=pos.index)
distances.head()
```

```python
most_similar = distances.max(axis=1).sort_values(ascending=False)
most_similar.head()
```

```python
similar_ids = distances.idxmax(axis=1).reindex(index=most_similar.index)
similar_ids.head()
```

```python
most_similar.hist()
```

```python
pos_distances = pd.DataFrame(cosine_similarity(pos.values, pos.values), index=pos.index, columns=pos.index)
pos_distances.head()
```

```python
pos_distances.describe()
```

```python
for i in range(0, len(pos_distances)):
    pos_distances.iloc[i,i] = 0
```

```python
pos_most_similar = pos_distances.max(axis=1).sort_values(ascending=False)
pos_most_similar.head()
```

```python
pos_most_similar.hist(bins=100)
```

```python
pos_most_similar.quantile(0.5)
```
